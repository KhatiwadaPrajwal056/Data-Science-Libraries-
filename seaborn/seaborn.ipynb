# Distribution plots

Basically for univariant set of observations and visualize it through a histogram.
Only one observation and hence we choose one column.

```python
sns.displot(car['speeding'],bins=25,kde=True)
plt.show()
```

# Joint plot

Relation between two variables and describe their distribution.
Compares two distribution and plots a scatter plot by default.


```python
sns.jointplot(data=car , x = 'speeding', y = 'alcohol',kind='reg')
plt.show()

# hex,reg,kde are hexagonal, regression, kernal densoty estimation
```

# KDE ( Kernal Density Estimation Plot)

Plots probability density function of the continuous or non parameteric data variables. i.e we can plot univariate or multi variables altogether

```python
sns.kdeplot(car['alcohol'])
plt.show()
```

# Pair plot

# Rug Plot

# Styling

```python
# darkgrid, whitegrid, dark , white, and ticks
sns.set_style('ticks')

# size

plt.figure(figsize=(10,10))

# joint plot
sns.jointplot(data=car, x = 'speeding' , y = 'alcohol', kind= 'reg')

#context option: ralk, paper, poster; used for different presentation
sns.set_context('paper', font_scale=1.2)

# remove axes
sns.despine(left=True, bottom=False)

plt.show()
```

# box plots

summary of data and outliers

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', palette='rocket')
plt.legend(loc=2)
plt.show()
```

# Violin Plot
Hybrid of box plot and KDE plot

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, palette='dark')
plt.show()
```

# strip plot
single axis scatter plot that gives distribution of many in 1D categorical values

```python
plt.figure(figsize=(10,5))
sns.stripplot(data=tips, x='day',y='total_bill', palette='dark',jitter=True, hue='sex',dodge=True)
plt.show()
```

# Heat map
2D representation of data representated bby color


```python
# for car crash
plt.figure(figsize=(10,8))
car = car.drop('abbrev',axis=1) # Droppping categorical column
sns.set_context('paper', font_scale = 1.2)
sns.heatmap(car.corr(), annot=True, cmap='Blues')
plt.show()
```

```python
# iris 

plt.figure(figsize=(10,8))
iris = iris.drop('species',axis=1) # Droppping categorical column
sns.set_context('paper', font_scale = 1.2)
sns.heatmap(iris.corr(), annot=True, cmap='Blues')
plt.show()
```

```python
# # flights
flight = flight.pivot_table(index='month',columns='year',values='passengers')
# flight = flight.drop('month',axis=1) # Droppping categorical column

sns.heatmap(data=flight, cmap='Purples', linecolor='white', linewidth = 0.5)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```

# pair grid

```python
iris = sns.load_dataset('iris')
iris_grid = sns.PairGrid(data=iris, hue='species')

iris_grid.map_diag(plt.hist)

iris_grid.map_upper(plt.scatter)
iris_grid.map_lower(sns.kdeplot)

plt.show()
```
