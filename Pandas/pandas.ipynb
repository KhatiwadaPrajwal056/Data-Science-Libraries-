#Working with the missing data:

```python
#The real dataset often have missing data.we use fill
car_sales=pd.read_csv("car-sales-missing-data.csv")
car_sales                                           #Nan=no value or not a number
```

```python
car_sales["Odometer"]=car_sales['Odometer'].fillna(car_sales['Odometer'].mean()) 
#either can be done by reassigning without inplace parameter
#fillna(value=None, *, method=None, axis=None, inplace=False, limit=None, downcast=None)                
#Needs to assign the filled value in the original csv file
#Not a good practice tho
#the value can't be a list
```

```python
#Initializing own index:
new_index=pd.Index(["a","b","c","d","e","f","g","h","i","j"])
car_sales_indexed=car_sales.set_index(new_index)
car_sales_indexed
```

```python
car_sales["Odometer"]=car_sales["Odometer"].round()
car_sales
```

```python
#Dropna()
#Either put inplace True or use reassgning
car_sales_droppped= car_sales.dropna()
car_sales_droppped
```

#Adding column using series in a existing dataset

for a series the no of value doesnot matter at most

for a list it must be exact number of values as per the index of the file

```python
#Column from series
# seats_column=pd.Series([1,2,3,4,5])
car_sales["Seats"]=pd.Series([1,2,3,4,5])
```

```python
car_sales   #Since the series is of only 5 items so remaining positions remains empty
```

```python
car_sales["Seats"]=car_sales["Seats"].fillna(car_sales["Seats"].mean())
car_sales
```

```python
#Column from list:
max_speed=[100,150,120,111,112,112,114,112,115,118] 
#While using a list, number of data should be as the index if the datas in the file 
car_sales["Speed"]=max_speed
car_sales
```

```python
#Using differernt columns to form different column:
car_sales["Seats per speed"]=car_sales["Seats"]/car_sales["Speed"]
car_sales
```

#Crosstab

```python
#Cross Tab
pd.crosstab(car_sales["Seats"],car_sales["Speed"])
```

#Removing columns:

```python
# using drop:
car_sales=car_sales.drop("Speed",axis=1)
car_sales
```

```python
#Using del:
del car_sales["Seats per speed"]
car_sales
```

#Shuffling and samling the data:

```python
#Shuffles the data each time differently
car_sales_shuffled=car_sales.sample(frac=1)
car_sales_shuffled
```

```python
#Sample fractions of data:
car_sales_shuffled.sample(frac=0.5) #Gives the 50% of the data also gets shuffled
```

```python
#Reseting the shuffled data:
car_sales_shuffled.reset_index(drop=True,inplace=True)  #Drop =do not add new index column(reset) in a data frame which is true
car_sales_shuffled
```
