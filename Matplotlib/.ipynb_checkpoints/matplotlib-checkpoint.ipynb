```python
x=[1,2,3]
y=[11,22,33]
plt.plot(x,y)
```

```python
# Use object oriented interfance rather than pyplot interface
```

# 1st method


```python
fig = plt.figure()   # creates figures
ax = fig.add_subplot()  #adds some axes
```

# 2nd method

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
```

# 3rd method : most used


```python
fig, ax = plt.subplots()
ax.plot(x,y)
```

Everytime we call fig its gonna reset the figure
