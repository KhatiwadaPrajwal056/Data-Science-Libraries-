# Matplotlib workflow example

```python
# 1. prepare the data
x=[1,2,3,4]
y=[10,10,30,40]

# 2. setup plot
fig,ax = plt.subplots(figsize=(10,5))  #width,height

# 3. plot data
ax.plot(x,y)
ax.set(title="Simple plot",
      xlabel="X-axis",
      ylabel="Y-axis")
plt.show()
```

# Bar plot : default vertical graph

```python
# make a dictionary and plot it
prices={"Book":10,
        "Pen":5,
        "Copy":7
       }
fig,ax=plt.subplots(figsize=(10,5))
ax.bar(prices.keys(),prices.values())
ax.set(title="Price Plot",
      xlabel="Materials",
      ylabel="Price");
```

```python
# Horizontal graph
prices={"Book":10,
        "Pen":5,
        "Copy":7
       }
fig,ax=plt.subplots(figsize=(10,5))
ax.barh(list(prices.keys()),list(prices.values()))
ax.set(title="Price Plot",
      xlabel="Materials",
      ylabel="Price");
```

# Histogram

```python
# Random array
x=np.random.randn(10000)
fig,ax=plt.subplots()
ax.hist(x);
```

# Subplots 1:

```python
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                        ncols=2,
                                        figsize=(10,10))
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(prices.keys(),prices.values())
ax4.hist(np.random.random(5));
```

# Subplots 2:


```python
fig, ax = plt.subplots(figsize=(10,5),
                      nrows=2,
                      ncols=2)
ax[0,0].plot(x,x/2);       #specifies the location i.e oth row oth column
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(prices.keys(),prices.values())
ax[1,1].hist(np.random.random(5));
```

# Plot same for car_sales

```python
#convert price to numeric:
car_sales["Price"]=car_sales["Price"].str.replace("[\$\,]",'')
car_sales
```

```python
# add sales column according to the date
car_sales["Sale Date"]=pd.date_range("1/1/2024",periods=10)
car_sales
```

```python
# Add total sale price using cumsum:
car_sales["Total sale"]=car_sales["Price"].cumsum()
car_sales
```

# Above it just concatenates the prices , because price is still str type

```python
#Resolve it:
car_sales["Total sales"]=car_sales["Price"].astype(int64).cumsum()
car_sales
```
