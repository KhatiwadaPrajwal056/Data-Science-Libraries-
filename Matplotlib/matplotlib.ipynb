# Bar plot : default vertical graph

```python
# make a dictionary and plot it
prices={"Book":10,
        "Pen":5,
        "Copy":7
       }
fig,ax=plt.subplots(figsize=(10,5))
ax.bar(prices.keys(),prices.values())
ax.set(title="Price Plot",
      xlabel="Materials",
      ylabel="Price");
```

```python
# Horizontal graph
prices={"Book":10,
        "Pen":5,
        "Copy":7
       }
fig,ax=plt.subplots(figsize=(10,5))
ax.barh(list(prices.keys()),list(prices.values()))
ax.set(title="Price Plot",
      xlabel="Materials",
      ylabel="Price");
```

# Which method to use?(Pyplot vs matplotlib OO method)
 Use pyplot plot for quick visualization but for better understanding use OO method

```python
heart_disease
```

```python
# Retrive the data of the people over age 50 
over50=heart_disease[heart_disease["age"]>50]
over50
```

# Pyplot method:

```python

#Scatter plot for the age and the cholestrol with target color
over50.plot.scatter(x="age",
            y="chol",
            c="target");
```

# OO method: scratch (Recommended)

```python
#OO method from scratch:
fig, ax= plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over50["age"],
           y=over50["chol"],
          c=over50["target"])

# make a labels:
ax.set(title="Heart disease and cholestrol level",
      xlabel="AGE",
      ylabel="CHOLESTROL")
# like the previous one there is a legend on the right side, lets put that too
ax.legend(*scatter.legend_elements(),title="target")
ax.set_xlim(50,75);  #Set the x-axis view limits.
```

# OO method (Alternative)

```python
fig, ax= plt.subplots(figsize=(10,6))
over50.plot(kind='scatter',
        x='age',
        y='chol',
        c='target',
        ax=ax)
ax.set_xlim(50,75);  #Set the x-axis view limits.
```

# Adding a horizontal inside of the plot

```python
fig, ax= plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over50["age"],
           y=over50["chol"],
          c=over50["target"])

# make a labels:
ax.set(title="Heart disease and cholestrol level",
      xlabel="AGE",
      ylabel="CHOLESTROL")
# like the previous one there is a legend on the left side, lets put that too
ax.legend(*scatter.legend_elements(),title="target")
ax.set_xlim(50,75) #Set the x-axis view limits.

# a horizontal line:
ax.axhline(250,xmin=0,xmax=1,linestyle="--"); 
# xmin to xmax: 0-1 i.e origin to max x_axis(can be made according to the need)
```

# Subplot of chol , thalach vs age:

```python
#creating axes:
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           #sharex=True 
                           )               #Note: here x axis is shared in both subplots: 
# adding data to ax0:
scatter=ax0.scatter(x=over50["age"],
                   y=over50["chol"],
                   c=over50['target'])
# customize:
ax0.set(title="Heart disease & Cholestrol level",
       xlabel="Age",
       ylabel="Cholestrol",
       )
# adding legend:
ax0.legend(*scatter.legend_elements(),title="target")
# mean line:
ax0.axhline(250,linestyle='--')

# adding data to ax1:
scatter1=ax1.scatter(x=over50["age"],
                    y=over50["thalach"],
                    c=over50['target'])

# customize:
ax1.set(title="Heart disease & Thalach level",
        xlabel="Age",
       ylabel="Thalach")

# adding legend:
ax1.legend(*scatter.legend_elements(),title="target")
# mean line:
ax1.axhline(140,linestyle='--')

fig.suptitle('Heart disease analysis',fontsize=15,fontweight='bold');

```
